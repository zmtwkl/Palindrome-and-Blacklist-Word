## Palindrome

### No. 1

```python
input = 'malam'
reverse = input[::-1]

if input == reverse:
  print(input + ' is a Palindrome')
else:
  print(input + ' is not a Palindrome')
```

## Black List Word

```python
black_list_word = ['andi', 'susi', 'bola']
```

### No. 1

```python
kalimat = 'andi sedang makan nasi padang'
kalimat = kalimat.split()
```

```python
if  black_list_word[0] in kalimat:
  print('ada black list word')
elif black_list_word[1] in kalimat:
  print('ada black list word')
elif  black_list_word[2] in kalimat:
  print('ada black list word')
else:
  print('tidak ada black list word')
```
